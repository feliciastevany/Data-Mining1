7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
label_encoder = LabelEncoder()
df_hero['role'] = label_encoder.fit_transform(df_hero['role'])
print(df_hero)

# Separate features and target variable
X = df_hero[['defense_overall',
              'offense_overall',
              'skill_effect_overall',
              'difficulty_overall',
              'movement_spd',
              'magic_defense',
              'mana',
              'hp_regen',
              'physical_atk',
              'physical_defense',
              'hp',
              'attack_speed',
              'mana_regen']]
y = df_hero['role']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

# Create a Decision Tree Classifier
dTreeClf = DecisionTreeClassifier()

# Train the model
dTreeClf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dTreeClf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy: {}".format(accuracy))

# Predict the role for the given character
new_character = {'defense_overall': 6.0,
                 'offense_overall': 7.0,
                 'skill_effect_overall': 4.0,
                 'difficulty_overall': 3.0,
                 'movement_spd': 250,
                 'magic_defense': 9,
                 'mana': 380,
                 'hp_regen': 37,
                 'physical_atk': 119,
                 'physical_defense': 24,
                 'hp': 2689,
                 'attack_speed': 0.75,
                 'mana_regen': 16.0 }

char_data = pd.DataFrame([new_character])
predicted_role = dTreeClf.predict(char_data)

# Get the predicted role (dicoding)
name_of_role = label_encoder.inverse_transform(predicted_role)

print(f"The predicted role for the character: {name_of_role[0]}")
```
